```python
#1098
#Asmaul Husna Aishe
features = [
    [5, 85], [1, 40], [4, 75], [2, 90],
    [6, 95], [3, 70], [2, 50], [0, 10],
    [4, 60], [5, 80], [1, 95], [3, 85]
]

labels = [1, 0, 1, 0, 1, 1, 0, 0, 0, 1, 0, 1]
```

```python
w1 = 0.1
w2 = 0.1
bias = 0
lr = 0.01

def predict(study, attendance):

    weighted_sum = (study * w1) + (attendance * w2) + bias

    #Step Function
    return 1 if weighted_sum >= 0 else 0
```

```python

epochs = 100

for epoch in range(epochs):
    total_error = 0


    for i in range(len(features)):


        student_data = features[i]
        x1 = student_data[0]
        x2 = student_data[1]
        target = labels[i]
        prediction = predict(x1, x2)


        error = target - prediction

        # If the model is wrong (error is not 0), we update the weights
        if error != 0:

            w1 = w1 + (lr * error * x1)


            w2 = w2 + (lr * error * x2)


            bias = bias + (lr * error)


            total_error = total_error + 1


    if (epoch + 1) == 20 or (epoch + 1) == 40 or (epoch + 1) == 60 or (epoch + 1) == 80 or (epoch + 1) == 100:
        print("Epoch number:")
        print(epoch + 1)
        print("Mistakes made:")
        print(total_error)
```

```python

study_val = input("Enter study hours: ")
study_val = float(study_val)

attend_val = input("Enter attendance percentage: ")
attend_val = float(attend_val)


prediction = predict(study_val, attend_val)


if prediction == 1:
    print("The student is likely to Pass")
else:
    print("The student is likely to Fail")
```
